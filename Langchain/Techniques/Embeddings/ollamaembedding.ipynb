Ollama supports embedding models ,making it possible to build retrieval augmented generation(RAG) application that combine text prompts with existing documents or other data

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    # we are using ollama 2b
    OllamaEmbeddings(model="gemma:2b")
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpka is the first letter of Greek alphabet",
        "Beta is the second letter of Greek alphabet"
    ]
)
```

```python
r1
```

```python
embeddings.embed_documents("What are the greek alphabets")
```

```python
## Other embedding models
### https://ollama.com/blog/emmbedding-models
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
text = "This is a test document."
query_results = embeddings.embed_query(text)
query_results
```
