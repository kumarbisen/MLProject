```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
#langsmith Tracking
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]= "true"
os.environ['LANGCHAIN_PROJECT']= os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
loader = WebBaseLoader("https://www.thehindu.com/news/national/operation-sindoor-was-a-reality-check-says-defence-secretary-india-us-defence-cooperation-remains-steady/article70042391.ece")
loader
```

```python
docs = loader.load()
docs
```

```python
# text splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap=200)
documents = text_splitter.split_documents(docs)
documents
```

```python
# convert into vector using openai embeddings
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings()
```

```python
from langchain_community.vectorstores import FAISS
vectorstoredb = FAISS.from_documents(documents,embeddings)
vectorstoredb
```

```python
query = "Defence Secretary Rajesh Kumar Singh said on Friday.Speaking at the inaugural STRIDE 2025 seminar "
result  = vectorstoredb.similarity_search(query)
result[0].page_content
```
